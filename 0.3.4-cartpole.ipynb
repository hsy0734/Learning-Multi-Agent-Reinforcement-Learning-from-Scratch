```python
import gymnasium as gym
import time
import pygame
import random
if __name__ == "__main__":
    pygame.init()
    env=gym.make("CartPole-v1",render_mode="human")
    state,_=env.reset()#初始化环境并获取初始状态，state是一个包含4个元素的数组，
                        #分别是小车的位置、速度、杆的角度、杆的尖端速度
                         #下划线_是环境返回的额外调试信息，这里我们不需要

    cart_position = state[0] # 小车的位置cart_speed = state[1]
    cart_velocity = state[1] # 小车的速度
    pole_angle = state[2] # 杆的角度
    pole_velocity = state[3] # 杆的尖端速度

    time.sleep(1)

    #打印状态
    print(f"state: {state}")
    print(f"cart_position: {cart_position}")
    print(f"cart_velocity: {cart_velocity}")
    print(f"pole_angle: {pole_angle}")
    print(f"pole_velocity: {pole_velocity}")

    #实现游戏过程
    start_time=time.time()#记录游戏开始时间
    max_steps=1000#最大步数
    step=0#当前步数
    failed=False#是否失败
    for step in range(1,max_steps+1):
        time.sleep(0.1)#每步休息0.1秒，方便人眼观察
        #根据键盘输入选择动作
        keys=pygame.key.get_pressed()
        action=0
        #如果没有收到任何键盘输入，则随机选择动作
        if not keys[pygame.K_LEFT] and not keys[pygame.K_RIGHT]:
            action=random.choice([0,1])
            #如果收到左键，向左推
        elif keys[pygame.K_LEFT]:
            action=0
        #如果收到右键，向右推
        elif keys[pygame.K_RIGHT]:
            action=1
        #控制环境中的小车
        state,_,done,_,_=env.step(action)
        #如果环境判断游戏失败（例如，小车超出屏幕边界），则结束游戏
        if done:
            failed=True
            break
            
        #打印当前状态
        print(f"step: {step}, action: {action}, state: {state}")
        #计算结束时间
        end_time=time.time()
        #打印游戏时长
        print(f"game duration: {end_time-start_time}")
        #打印游戏时间和步数结果
        print(f"game time: {end_time-start_time}, steps: {step}")

```
